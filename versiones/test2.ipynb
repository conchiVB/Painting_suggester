```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
path = './data/'
```

```python
img = cv2.imread(path + 'test01.jpg')
#img= img[90:400,20:540]
plt.imshow(img)
plt.show()

```

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.show()

#purple
rgb_color1=np.asarray([51, 0, 102])
rgb_color2=np.asarray([229, 100, 255])
mask=cv2.inRange(img_rgb,rgb_color1, rgb_color2)
plt.imshow(mask, cmap='gray')
plt.show()

```

```python
#green
plt.imshow(img_rgb)
plt.show()

rgb_color1=np.asarray([0, 100, 51])
rgb_color2=np.asarray([100, 255, 204])
mask=cv2.inRange(img_rgb,rgb_color1, rgb_color2)
plt.imshow(mask, cmap='gray')
plt.show()
```
