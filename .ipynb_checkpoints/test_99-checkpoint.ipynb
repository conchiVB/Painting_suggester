### Color Separation in an Image using KMeans Clustering using Python
https://medium.com/analytics-vidhya/color-separation-in-an-image-using-kmeans-clustering-using-python-f994fa398454


```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2
from sklearn.cluster import KMeans
```

```python
#img=cv2.imread('image color2.jpg')
img = cv2.imread('./data/DH/DH01.jpg')
plt.imshow(img)
```

```python
img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
img=img.reshape((img.shape[1]*img.shape[0],3))
```

```python
kmeans=KMeans(n_clusters=24)
s=kmeans.fit(img)
```

```python

labels=kmeans.labels_
print(labels)
labels=list(labels)
```

```python
centroid=kmeans.cluster_centers_
print(centroid)
```

```python
percent=[]
for i in range(len(centroid)):
    j=labels.count(i)
    j=j/(len(labels))
    percent.append(j)
print(percent)
```

```python
plt.pie(percent,colors=np.array(centroid/255),labels=np.arange(len(centroid)))
plt.show()
```

```python
md=[]
for i in range(1,40):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(img)
    o=kmeans.inertia_
    md.append(o)
print(md)
```

```python

plt.plot(list(np.arange(1,40)),md)
plt.show()
```
